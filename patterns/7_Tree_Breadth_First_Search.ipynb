```python
from collections import deque

class TreeNode:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

def averageByLevel(root):
  result = []

  if root is None:
    return result

  d = deque()
  d.append(root)
  while d:
    levelSum = 0
    levelSize = len(d)
    for _ in range(levelSize):
      node = d.popleft()
      levelSum += node.data

      if node.left:
        d.append(node.left)

      if node.right:
        d.append(node.right)

    result.append(levelSum/levelSize)


  return result

def main():
  root = TreeNode(1)
  root.left = TreeNode(2)
  root.left.left = TreeNode(4)
  root.left.right = TreeNode(5)
  root.right = TreeNode(3)
  root.right.left = TreeNode(6)
  root.right.right = TreeNode(7)

  print("Output: " + str(averageByLevel(root)))

  root = TreeNode(12)
  root.left = TreeNode(7)
  root.left.left = TreeNode(9)
  root.left.right = TreeNode(2)
  root.right = TreeNode(1)
  root.right.left = TreeNode(10)
  root.right.left.left = TreeNode(20)
  root.right.left.right = TreeNode(17)
  root.right.right = TreeNode(5)

  print("Output: " + str(averageByLevel(root)))

main()
```

```python
from collections import deque

class TreeNode:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

def getSuccessor(root, givenNode):
  if root is None:
    return None

  d = deque()
  d.append(root)

  while d:
    node = d.popleft()

    if node.left:
      d.append(node.left)

    if node.right:
      d.append(node.right)

    if node == givenNode:
      break

  return d.popleft().data if d else None

def main():
  root = TreeNode(1)
  root.left = TreeNode(2)
  root.left.left = TreeNode(4)
  root.left.right = TreeNode(5)
  root.right = TreeNode(3)
  root.right.left = TreeNode(6)
  root.right.right = TreeNode(7)

  print("Output: " + str(getSuccessor(root, root.right)))

  root = TreeNode(12)
  root.left = TreeNode(7)
  root.left.left = TreeNode(9)
  root.right = TreeNode(1)
  root.right.left = TreeNode(10)
  root.right.left.left = TreeNode(20)
  root.right.left.right = TreeNode(17)
  root.right.right = TreeNode(5)

  print("Output: " + str(getSuccessor(root, root.left.left)))

main()
```

```python
from collections import deque

class TreeNode:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

def getMinimumDepth(root):
  depth = 0

  if root is None:
    return 0

  d = deque()
  d.append(root)
  while d:
    levelSize = len(d)
    depth += 1
    for _ in range(levelSize):
      node = d.popleft()

      if not node.left and not node.right:
        return depth

      if node.left:
        d.append(node.left)

      if node.right:
        d.append(node.right)


def main():
  root = TreeNode(1)
  root.left = TreeNode(2)
  root.left.left = TreeNode(4)
  root.left.right = TreeNode(5)
  root.right = TreeNode(3)

  print("Output: " + str(getMinimumDepth(root)))

  root = TreeNode(12)
  root.left = TreeNode(7)
  root.right = TreeNode(1)
  root.right.left = TreeNode(10)
  root.right.right = TreeNode(5)

  print("Output: " + str(getMinimumDepth(root)))

  root = TreeNode(12)
  root.left = TreeNode(7)
  root.left.left = TreeNode(9)
  root.right = TreeNode(1)
  root.right.left = TreeNode(10)
  root.right.left.left = TreeNode(20)
  root.right.left.right = TreeNode(17)
  root.right.right = TreeNode(5)

  print("Output: " + str(getMinimumDepth(root)))

main()
```

```python
from collections import deque

class TreeNode:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None
    self.successor = None

  def __repr__(self):
    result = []

    leftestFromLevel = self
    while leftestFromLevel:
      node = leftestFromLevel
      leftestFromLevel = None
      currLevelResult = []
      while node:
        currLevelResult.append(node.data)

        if leftestFromLevel is None:
          if node.left:
            leftestFromLevel = node.left
          elif node.right:
            leftestFromLevel = node.right

        node = node.successor

      result.append(currLevelResult)

    return str(result)

def connectSuccessor(root):
  if root is None:
    return

  d = deque()
  d.append(root)
  while d:
    levelSize = len(d)
    prev = None
    for _ in range(levelSize):
      node = d.popleft()
      if prev:
        prev.successor = node
      prev = node

      if node.left:
        d.append(node.left)

      if node.right:
        d.append(node.right)


def main():
  root = TreeNode(1)
  root.left = TreeNode(2)
  root.left.left = TreeNode(4)
  root.left.right = TreeNode(5)
  root.right = TreeNode(3)
  root.right.left = TreeNode(6)
  root.right.right = TreeNode(7)

  print("Output: " + str(connectSuccessor(root)))
  print(root)

  root = TreeNode(12)
  root.left = TreeNode(7)
  root.left.left = TreeNode(9)
  root.right = TreeNode(1)
  root.right.left = TreeNode(10)
  root.right.right = TreeNode(5)

  print("Output: " + str(connectSuccessor(root)))
  print(root)

main()
```

```python
def removeSpaces(string):
  L = 0
  arr = list(string)

  if len(arr) > 0:
    arr[0] = arr[0].lower()

  for R in range(len(arr)):
    if arr[R] != " ":
      arr[L] = arr[R]
      L += 1
    elif R < len(arr) - 1:
      arr[R+1] = arr[R+1].upper()

  return "".join(arr[:L])


def main():
  print("Output: " +  str(removeSpaces("Reverse a Sub-list")))

main()
```
